```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
```

## Reading the image in grayscale

```python
original_image = cv2.imread("image.jpeg")
original_image = cv2.resize(original_image, (800, 350))
img = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)
cv2.imshow("image", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Converting an image to binary format

```python
T, tresh = cv2.threshold(img, 100, 255, cv2.THRESH_BINARY_INV)
tresh
```

```python
cv2.imshow("image", tresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cnts, _ = cv2.findContours(tresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
len(cnts)
```

```python
copy_image = original_image.copy()
for i in range(len(cnts)):
    cv2.drawContours(copy_image, cnts, i, (0, 255, 0), 3)
    x, y, w, h = cv2.boundingRect(cnts[i])
    cv2.rectangle(copy_image, (x-10, y-10), (x+w+10, y+h+10), (0, 255, 0), 3)
    cv2.imshow("image", copy_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
model = load_model("digit_classifier_CNN.h5")
for i in range(len(cnts)):
    x, y, w, h = cv2.boundingRect(cnts[i])
    extract_image = original_image[y-10: y+h+10, x-10: x+w+10]
    extract_image = cv2.resize(extract_image, (32, 32))
    extract_image = extract_image / 255.0
    extract_image = np.array([extract_image])

    output = model.predict(extract_image)[0]
    print(output)
    max_index = np.argmax(output)
    print(max_index)
    cv2.rectangle(original_image, (x-10, y-10), (x+w+10, y+h+10), (0, 255, 0), 3)
    cv2.putText(original_image, str(max_index), (x, y-20), cv2.FONT_HERSHEY_COMPLEX, 0.95,  (0, 255, 0), 3)

cv2.imshow("image", original_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
